```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

## Heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
tips_corr= tips[['total_bill','size','tip']]
```

```python
tips_corr
```

```python
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr(),annot=True)
```

## Cluster map

```python
sns.clustermap(tips_corr.corr(),annot=True)
```

## Pivot table Heat map

```python
pvtflight = flights.pivot_table(values = 'passengers',index = 'month',columns = 'year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```
